```python
# 파이썬은 #을 주석 기호로 사용합니다.
# 숫자 자료형 : 정수, 실수, 복소수, 8진수, 16진수
# 파이썬의 경우 저장하는 데이터의 형식에 따라 자료형이 설정됩니다.
a = 100 # a변수에 100을 저장 => 정수
b = 10.5 # b변수에 10.5를 저장 => 실수
c = 0o12 # c변수에 8진수 12를 저장 => 8진수
d = 0xA # d변수에 16진수 A를 저장 => 16진수
print(a)
print(b)
# 8진수, 16진수는 10진수로 변경하여 출력됩니다.
print(c)
print(d)
print(a + b + c + d) # 10진수 실수형으로 출력됩니다.
```

```python
# 숫자 연산자
# 사칙 연산 : +, -, *, /, %(나머지 연산)
# 제곱 연산자 : **
print(2**2)
print(2**32)
print(2**64)
# 소수점을 버리는 나누기 연산자, 몫만 출력하는 연산자 : //
print(7/3)
print(7//3)
```

```python
# 문자열 자료형 : 큰따옴표, 작은따옴표 상관없이 문자열을 저장할 수 있습니다.
# 역슬래시(\)를 사용하여 문자열 안에 따옴표를 문자로서 사용 가능합니다.
a = "'큰따옴표' 사용\""
b = '"작은따옴표" 사용'
# 문자열의 경우 엔터 사용 시 에러가 나지 않도록 하는 문자열 작성 방식
c = """큰따옴표
3개 사용"""
d = '''작은따옴표 3개
사용'''
print(c)
print(d)
```

```python
# 문자열 곱하기
a = "Python"
print(a*2)
# 문자열 인덱싱 : 문자열의 인덱스를 이용하여 특정 위치의 문자열을 출력하는 방식
print(a[1]) # 앞에서부터 1번째 문자를 출력
print(a[5]) # 앞에서부터 5번째 문자를 출력
print(a[-1]) # 뒤에서부터 1번째 문자를 출력
print(a[-5]) # 뒤에서부터 5번째 문자를 출력
```

```python
# 문자열 슬라이싱 : 특정 문자열을 잘라내어 출력하는 방식
a = "python is fun!"
print(a[5:7+1])
print(a[7:8+1])
# 왼쪽을 생략하는 슬라이싱
print(a[:7]) # 처음부터 7번 문자열까지 출력
print(a[:-1]) # 처음부터 뒤에서 1번 문자열까지 출력
# 오른쪽을 생략하는 슬라이싱
print(a[7:]) # 7번부터 마지막까지 출력
print(a[-5:]) # 뒤에서 5번재부터 마지막까지 출력
```

```python
# 문자열에서 사용 가능한 함수
a = "python is FUN! fun! FUN! fun! FUN!"
print(a.count("fun")) # 알파벳 p의 개수를 출력하는 함수
print(a.find("안녕")) # is가 몇 번째에 있는지 출력하는 함수, 없으면 -1 출력
print(a.index("fun")) # find와 같이 몇 번재에 있는지 출력, 없으면 에러 발생
print(a.upper()) # 소문자를 대문자로 변경
print(a.lower()) # 대문자를 소문자로 변경
```

```python
# 문자열 함수
a = "python"
print(",".join(a)) # 배열의 사이에 ,를 넣고 하나의 문자열로 연결
a = "java is good"
print(a.replace("java", "python"))
```

```python
# 리스트 자료형
a = [5, 10, 15, 20]
b = [2, 1.5, "홍길동", "이순신"]
print(a)
print(b)
# 리스트 인덱싱
print(a[0])
print(a[3])
print(b[-1])
print(b[-3])
# 리스트 슬라이싱
print(b[2:4])
```

```python
# 리스트의 연산자
a = [2, 4]
b = [5, 10]
print(a + b) # 더하기 연산 : 리스트끼리 더하여 새로운 리스트를 생성
print(a * 3) # 곱하기 연산 : 리스트를 세 번 곱하여 새로운 리스트를 생성
```

```python
# 리스트에서 사용하는 함수
a = [2, 4, 6, 1, 5, -10, 50, 32]
a.append(8) # 리스트에 데이터 추가 함수
print(a)
a.reverse() # 리스트의 순서를 반대로 변경
print(a)
a.sort() # 데이터 오름차순 정렬
print(a) 
a.sort(reverse = True) # 데이터 내림차순 정렬
print(a)
a.remove(50) # 특정 데이터를 삭제하는 함수
print(a)
print(a.pop(0)) # 인덱스를 통해 데이터를 출력 후 삭제하는 함수
print(a)
```

```python
# 튜플 자료형 : 리스트와 거의 같으나 내용을 변경할 수 없습니다.
a = (1, 2, 3, 4, 5)
print(a)
# +, * 연산, 인덱싱, 슬라이싱은 가능합니다.
b = (6, 7)
print(a + b)
print(b * 3)
```

```python
# 딕셔너리 자료형 : { key : value, key : value } 형식으로 만들어진 자료형
a = {'name' : '홍길동', 'phone' : '010-1234-5678', 'age' : 20, 'birth' : '0814'}
# value 값 출력하기
print(a['name'])
print(a['birth'])
# 데이터 추가하기
a['email'] = 'hong@gmail.com'
print(a)
# 데이터 삭제하기
del a['age']
print(a)
# 키 값을 리스트로 출력
print(list(a.keys()))
# value를 리스트로 출력
print(list(a.values()))
```

```python
# 집합 자료형
a = {1, 2, 4, 'a', 5, 1, 2} # 중복된 데이터는 저장되지 않고 하나만 남습니다.
b = {4, 5, 6, 7, 8, 9, 4, 5}
print(a)
print(b)
# 집합 자료형의 연산자
# & : 교집합 연산자
print(a & b)
print(a.intersection(b))
# | : 합집합 연산자
print(a | b)
print(a.union(b))
# - : 차집합 연산자
print(a - b)
print(a.difference(b))
# 데이터 한 개 추가
a.add(100)
print(a)
# 데이터 삭제
a.remove('a')
print(a)
# 데이터 여러개 추가
a.update([200, 300, 400, 500])
print(a)
```
