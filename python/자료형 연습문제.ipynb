```python
# [문제8] 주어진 리스트 [1, 2, 3, 4, 5]에서 첫 번째 원소와 마지막 원소를 교환하는 파이썬 코드를 작성하세요.

# 자바, 파이썬 둘 다 가능한 방식
a = [1, 2, 3, 4, 5]
temp = a[0]
a[0] = a[4]
a[4] = temp
print(a)

# 파이썬에서만 가능한 방식
a = [1, 2, 3, 4, 5]
a[0], a[4] = a[4], a[0]
print(a)
```

```python
# [문제9] 주어진 리스트 [5, 2, 7, 1, 9]를 오름차순 정렬하는 파이썬 코드를 작성하세요.

a = [5, 2, 7, 1, 9]
a.sort()
print(a)
```

```python
# [문제10] 주어진 리스트 [1, 3, 5, 4, 2]를 내림차순으로 정렬하는 파이썬 코드를 작성하세요.

a = [1, 3, 5, 4, 2]
a.sort(reverse = True)
print(a)
```

```python
# [문제15] 주어진 두 집합 {1, 2, 3}과 {2, 3, 4}의 교집합을 구하는 파이썬 코드를 작성하세요.

a = {1, 2, 3}
b = {2, 3, 4}
print(a & b)
print(a.intersection(b))
```
